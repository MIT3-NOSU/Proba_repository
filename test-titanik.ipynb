Проверить гипотезу о зависимости выживаемости от класса проживания

```python
#всего выжиших
surv = data['Survived'].value_counts()
surv
```

```python
# выживших в каждом классе
u = data['Survived'] == 1
survData = data[u]['Pclass'].value_counts()
survData
```

```python
# процент выживших в каждом классе к общему числу выживших
survData/surv[1]
```
